```python
def phenotype( pA, pB ):
    qA = 1 - pA
    qB = 1 - pB
    mut= qA ** 2 + qB ** 2 - ( qA ** 2 * qB ** 2 )
    WT = 1 - mut
    return WT

def allele_freq( locus ):
    sum = 0
    for i in locus:
        if i == 'A' or i == 'B':
            sum += 1
    p = sum / float(len(locus))
    return p
```

```python
import os
import csv
import numpy as np

os.chdir("/Users/jamessantangelo/Documents/Academia/Doctorate_PhD/Projects/SEC_Simulating.evolutionary.clines/SEC_Visuals/HCN-x-Alleles plot")

freqA = np.linspace(0, 1, num = 101)
# freqB = np.linspace(0, 1, num = 101)

with open("Homozygote-Heterozygote_HWE.csv", "wb") as f:
    writer = csv.writer(f)
    writer.writerow(["p", "q", "psquared_2pq"])

    for i in freqA:
        writer.writerow( [ i, 1 - i, round( i ** 2 + (2 * i * (1 - i)), 4 ) ] )
```
